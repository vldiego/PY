# Compresion de datos supervisada mediante análisis discriminante lineal

El análisis de discriminante lineal puede ser utilizado como una técnica para la extracción de características para aumentar la eficiencia computacional y reducir el grado de sobreajuste debido a la maldición de la dimensionalidad en modelos no regularizado.

El concepto general que hay detrás del ADL es muy parecido al del ACP. Si el ACP intenta encontrar los ejes del componente octogonal de varianza máxima en un conjunto de datos, el objetivo en el ADL es encontrar el subespacio de características que optimice la divisibilidad de clases. En las secciones siguientes, trataremos las similitudes entre el ADL y el ACP con mayor detalle y entraremos en el enfoque del ADL paso a paso.

## Análisis de componentes principales frente a análisis de discrimante lineal

Tanto el ADL como el ACP son técnicas de transformación lineal que pueden utilizarse para reducir el numero de dimensiones en un conjunto de datos; el primero es un algoritmo sin supervisión mientras que el segundo es supervisado. Por lo tanto, intuitivamente podríamos pensar que el ADL es una técnica de extracción de características superior para tareas de clasificación si lo comparamos con el ACP. Sin embargo, A. M. Martinez afirmo que  el preprocesamiento tiende a dar mejores resultados de clasificación en una tarea de reconocimiento de imágenes en determinados casos; por ejemplo, si cada clase cuenta con solo una pequeña cantidad de muestras.

## El funcionamiento interno del análisis discriminante lineal

1.	Estandarizar el conjunto de datos de d dimensiones (d es el número de características)

2.	Para cada clase, calcular el vector medio de d dimensiones.

3.	Construir la matriz de dispersión entre la clase Sb y la matriz de dispersión dentro de la clase $S_w$.

4.	 Calcular los autovectores y los correspondientes autovalores de la matriz $S_w^1$ $S_b$.

5.	Ordenar los autovalores en orden decreciente para clasificar los correspondientes autovectores.

6.	Elegir lo k autovectores que correspondan a los k valores mas altos para construir una matriz de transformación W de d x k dimensiones; los autovectores son las columnas de esa matriz.

7.	Proyectar las muestras en el nuevo subespacio de características mediante la matriz de transformación.
